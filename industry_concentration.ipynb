### Is the concentration of jobs across industries in an area correlated with wage levels for the local population?

The goal here is to demonstrate some key features of Polars, including:
- Lazy evaluation
- Query engine
- Speed
- Plotting

```python
import polars as pl
from polars_demo.utils import polars_hhi, LodesProd
from polars_demo.polars_flow import employment_hhi, acs_tracts, joined_df
```

```python
wac_df: pl.LazyFrame = employment_hhi(LodesProd.WAC)
# Check out the type of object
# What does it return?
# How to materialize results if you want?
```

```python
joined_df.collect().plot.scatter(x="median_inc_clean", y="wac_hhi", alpha=0.1)
```
